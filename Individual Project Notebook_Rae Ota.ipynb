## Research Question
My research question examines if there is any statistically significant correlation between the number of documented Covid-19 *<ins>cases</ins>* in 2020 and the top ten countries considered to have the most individualistic- or collectivist valuing societies. I aim to see if documented Covid-19 case numbers *increase or decrease consistently* depending on if the country and its people operate and value the individual as opposed to the collective, or vice versa.

**Collectivism vs. Individualism - what's the difference?**

Taken from the World Population Review, I believe the definition of collectivism and individualism are accurately described in the following passage: "Collectivism and individualism describe two values that lie on a spectrum apart from each other, and these two values have an immense impact on a country’s culture. Many Western countries tend towards individualism, as they value individual liberty and freedom and see it as the key to human flourishing. However, other countries see the individual as embedded within social relationships and appreciate the collective well-being of the group over the rights of the individual. Cultures in non-Western countries, such as South America, Central America, Asia, and Africa, tend towards collectivism and the well-being of the interconnected web of social relationships over the unbridled rights of the individual" ("Collectivist").

Photo by: https://byuipt.net

While the lines can get blurry, it is important to note that individualism and collectivism are not synonymous with authoritarianism or communism, since individualism and collectivism resemble cultural values more than hard-and-fast political ideologies or social systems. Countries like China, North Korea, and Russia have historically fallen in line with more authoritarianism-leaning or communist-leaning social structures, with having state ownership, public ownership, and following the authority of the government over one's personal beliefs and freedoms emphasized. I will not be including these countries in my data analysis because I want to see the effects of valuing community vs. the individual within a culture on the number of Covid-19 cases reported in that culture's country absent of external authorities that might have control of its people, their choices, and their actions, due to the power they can hold (Cherry). 

## My Hypothesis

I expect the data to reveal that at the beginning of the pandemic, the impact of group versus individual will be seen, with countries like Japan having lower Covid-19 cases compared to that of the United States. However, as borders start to open up (for instance, Japan had closed its borders to foreigners for a long time, but when they started opening up, there were outbreaks), there will be a more-or-less equal ratio of Covid-19 cases to country population, even as Covid-19 vaccines mitigate high increases in Covid-19 cases. Whether or not one has been raised in a country that values the collective above the individual, I think I will find results that demonstrate a worldwide disease will still find its way to a country’s population in some form or another — even if actively adhering to preventative policies can slow the spread of a pandemic, which is still a notable find.

**Potential bias/subjectivity in data** 

While there are no countries that are in absolute one-hundred percent collectivist or one-hundred percent individualistic, I have used different websites, as well as the *Hofstede Country Comparison Tool* which uses the Hofstede’s Cultural Dimensions Theory, to assess which countries most lean towards collectivism or individualism. This theory "is a framework used to understand the differences in culture across countries... to distinguish between different national cultures, the dimensions of culture, their impact on etiquette and to facilitate communication in areas ranging from business to diplomacy" (Wale).

This theory was pieced together by Geert Hofstede, a Dutch researcher, in 1980. He used many surveys during "the 1960s and 1970s" to "investigate variations in values within different sectors of IBM, a global computer manufacturing company. The study comprised over 100,000 employees from 50 countries across three regions" (Wale).

Still, it is good to keep in mind that there exists subjectivity in my choices of which countries are collectivist or individualist- valuing, despite using different sources to come to my final conclusions.
